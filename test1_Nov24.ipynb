```python
from fastai.vision.all import *
from PIL import Image

# Define the path to your image
image_path = Path(r"C:\Users\Geetansh\Desktop\Life\datatset\2024 (22)  - Copy.jpg")

# Open the image using PIL
img = PILImage.create(image_path)  # Convert to a format FastAI recognizes

# Resize the image to desired dimensions (e.g., 100x75 pixels)
resize_transform = Resize((75, 100), method='squish')  # 'squish' resizes without maintaining aspect ratio
resized_img = resize_transform(img)

# Display the resized image
show_image(resized_img, title="Resized Image")
```

```python
from torchvision.transforms.functional import to_tensor, to_pil_image

# Convert PILImage to tensor
resized_img_tensor = resize_transform(to_tensor(img))

# Convert tensor back to PIL image
resized_pil_img = to_pil_image(resized_img_tensor)

# Save the resized image to disk
output_path = Path(r"C:\Users\Geetansh\Desktop\Life\output\resized.jpg")
resized_pil_img.save(output_path)

print(f"Resized image saved to: {output_path}")
```
